Data Wrangling

```python
#dropping columns Unnamed:0 and id
df.drop(["Unnamed: 0","id"], axis = 1, inplace = True)
```

```python
#print the number of NaN values in columns bedrooms and bathrooms
print("Number of NaN values in column 'bedrooms' is:", df['bedrooms'].isnull().sum())
print("Number of NaN values in column 'bathrooms' is:",df['bathrooms'].isnull().sum())
```

```python
mean = df['bedrooms'].mean()
df['bedrooms'].replace(np.nan, mean,inplace = True )
```

```python
mean = df['bathrooms'].mean()
df['bathrooms'].replace(np.nan, mean,inplace = True )
```

Model Development

```python
#Fitting a linear regression model to predict the 'price' using the feature 'sqft_living' and calculating R^2
X1 = df[['sqft_living']]
Y1 = df[['price']]
lm1 = LinearRegression()
lm1.fit(X1,Y1)
lm1.score(X1,Y1)
```

```python
#Fit a linear regression model to predict the 'price' using the list of features and calulate R^2
features =["floors", "waterfront","lat" ,"bedrooms" ,"sqft_basement" ,"view" ,"bathrooms","sqft_living15","sqft_above","grade","sqft_living"]
X2 = df[features]
Y2 = df[['price']]
lm2 = LinearRegression()
lm2.fit(X2,Y2)
lm2.score(X2,Y2)
```

```python
#Use the list to create a pipeline object to predict the 'price', fit the object using the features in the list features, and calculate the R^2.
#ypipe is predicted price
Input=[('scale',StandardScaler()),('polynomial', PolynomialFeatures(include_bias=False)),('model',LinearRegression())]
pipe = Pipeline(Input)
pipe.fit(X2,Y2)
ypipe = pipe.predict(X2)
ypipe[0:10]
pipe.score(X2,Y2)
```
